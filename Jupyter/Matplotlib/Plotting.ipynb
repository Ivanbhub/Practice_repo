```python
#plotting, adding [markers,colors, linestyle] and labels 

plt.plot(ages_x,dev_y, color = 'c', linestyle='--',marker='.',label='All developers')
plt.xlabel('Age')
plt.ylabel('Median Salary($)')
plt.title('Median Salary ($) by Age')
plt.plot(ages_x,py_dev_y, '^r-',label='Python devs')


# adding legends

plt.legend()
plt.show()
```

```python
#plotting, adding color with HEX values and labels 



plt.plot(ages_x,dev_y, color = '#031cfc', linestyle='--',label='All developers')
plt.xlabel('Age')
plt.ylabel('Median Salary($)')
plt.title('Median Salary ($) by Age')
plt.plot(ages_x,py_dev_y, color='#03fc62',label='Python devs')


# adding legends

plt.legend()
plt.show()
```

```python
# added a 3rd line for comparison
# changed linewidth 


plt.xlabel('Age')
plt.ylabel('Median Salary($)')
plt.title('Median Salary ($) by Age')
plt.plot(ages_x,dev_y, color = '#031cfc', linestyle='--',linewidth='3',label='All developers')
plt.plot(ages_x,py_dev_y, color='#03fc62',linewidth='2',label='Python devs')
plt.plot(ages_x,js_dev_y, color='#e3001a',linewidth='2', label='JavaScript devs')


# adding legends

plt.legend()

plt.show()
```

```python
# added a grid to make it easier to read 


plt.xlabel('Age')
plt.ylabel('Median Salary($)')
plt.title('Median Salary ($) by Age')
plt.plot(ages_x,dev_y, color = '#031cfc', linestyle='--',linewidth='3',label='All developers')
plt.plot(ages_x,py_dev_y, color='#03fc62',linewidth='2',label='Python devs')
plt.plot(ages_x,js_dev_y, color='#e3001a',linewidth='2', label='JavaScript devs')

plt.legend()

plt.grid(True)

plt.show()
```
