```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('ggplot')
plt.title('My Pie Chart')
slices = [40,20]
labels=['forty','twenty']
plt.pie(slices,labels=labels)
plt.show()
```

```python
# changed the edge color to black:

plt.style.use('ggplot')
plt.title('My Pie Chart')
slices = [40,20]
labels=['forty','twenty']
plt.pie(slices,labels=labels, wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
#added more slices and changed color using hex

plt.style.use('ggplot')
plt.title('My Pie Chart')
slices = [40,20,30,60]
labels=['forty','twenty','xx','yy']
colors=['#5e7cad','#7bb399','#a188b5','#ccb697']
plt.pie(slices,labels=labels,colors=colors, wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.style.use('ggplot')
plt.title('Language Popularity')
# emphasized the Pyhton slice with explode
explode = [0,0,0,.1,0]

plt.pie(slices,labels=labels,explode=explode, wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.style.use('ggplot')
plt.title('Language Popularity')

explode = [0,0,0,.1,0]

# added a shadow and changed the starting angle 
plt.pie(slices,labels=labels,explode=explode, shadow= True, startangle=80, wedgeprops={'edgecolor': 'black'})
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.style.use('ggplot')
plt.title('Language Popularity')

explode = [0,0,0,.1,0]

# added the percentage of each slice using the percentage format string %1.1f%%

plt.pie(slices,labels=labels,explode=explode, shadow= True, startangle=80, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})
plt.show()

```
