```python
# Scatter Plots

from matplotlib import pyplot as plt
```

```python
# My Data

x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
# Style of plots

plt.style.use('seaborn')
```

```python
plt.scatter(x,y)

plt.show()
```

```python
# Looks like there is no correlation in my data
```

```python
# increased the size (s) of the dots and color(c) :

plt.scatter(x,y, s= 300, c='brown')

plt.show()
```

```python
# changed the marker :

plt.scatter(x,y, s= 100, c='brown', marker='x')

plt.show()
```

```python
# Changed the the edgecolor & size and increased transparency using alpha:

plt.scatter(x,y, s= 100, c='brown', edgecolor='black',linewidth=1, alpha=0.70)

plt.show()
```

```python
# added a 3rd variable for comparison 
# the number corespond to the x and y defined above

rating = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
```

```python
# rating will determine the intensity of the color. The closer to 0 the lighter it is. 

plt.scatter(x,y, s= 100, c=rating,cmap='Reds', edgecolor='black',linewidth=1, alpha=0.70)

plt.show()
```

```python
# added a label for viewer to understand the plot:

plt.scatter(x,y, s= 100, c=rating,cmap='Reds', edgecolor='black',linewidth=1, alpha=0.70)

# added a label for viewer to understand the plot:

cbar =plt.colorbar()
cbar.set_label('Rating')


plt.show()
```

```python
# added a a 4th variable (sizes) for comparison. the bigger the number the bigger the circle
# the numbers in the variable sizes corespond to the x and y defined above

rating = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]



plt.scatter(x,y, s= sizes, c=rating,cmap='Reds', edgecolor='black',linewidth=1, alpha=0.70)



cbar =plt.colorbar()
cbar.set_label('Rating')


plt.show()
```

```python
import pandas as pd
```

```python
# My Data

data = pd.read_csv('Data/YT_likes.csv')
```

```python
 data  #top 200 trending videos 
```

```python
# Question: Is there a correlation between view count and likes
```

```python
views=data['view_count']
likes=data['likes']
ratio=data['ratio']
```

```python


plt.scatter(views,likes, edgecolor='black',linewidth=1, alpha=0.70)



plt.title('Trending Youtube Video')
plt.xlabel('Views')
plt.ylabel('Total Likes')

plt.show()
```

```python
# looks like there is an outlier. Using log scales would prevent the outlier from skewing our scatter plot.
```

```python
plt.scatter(views,likes, edgecolor='black',linewidth=1, alpha=0.70)

#added log scales

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Video')
plt.xlabel('Views')
plt.ylabel('Total Likes')

plt.show()
```

```python
# the ratio of likes and dislikes can be a good metric to use for the color of our points 
```

```python
# added c for color 

plt.scatter(views,likes,c = ratio,cmap='autumn', edgecolor='black',linewidth=1, alpha=0.70)

# added a color bar for description

cbar = plt.colorbar()
cbar.set_label('Like/Dislike ratio')

plt.xscale('log')
plt.yscale('log')

plt.title('Trending Youtube Video')
plt.xlabel('Views')
plt.ylabel('Total Likes')

plt.show()
```
