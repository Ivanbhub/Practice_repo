```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = pd.read_csv('data/data_for_lang_worked_with.csv')
```

```python
#the first 10 of my data
data.head(10)
```

```python
# Defining my variables

age = data['Age']

all_devs = data['All_Devs']

py = data['Python']

jv = data['JavaScript']
```

```python
# subplotting

# 1 row and 1 column

fig1, ax1= plt.subplots()


ax1.plot(age, all_devs, label='All devs salary')

ax1.plot(age, py, label='Python dev salary')

ax1.plot(age, jv, linestyle='--', label='JavaScript Salary')

ax1.set_title('Salary Comparison')
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.legend()
plt.show()
```

```python
########################### this will help you understand how subplots work : ######################
# 1 row and 1 column
fig1, ax1= plt.subplots(nrows=1,ncols=1)
```

```python
# 2 rows and 1 column
fig1, ax1= plt.subplots(nrows=2,ncols=1)
```

```python
# 3 rows and 1 column
fig1, ax1= plt.subplots(nrows=3,ncols=1)
```

```python
# 3 rows and 2 columns
fig1, ax1= plt.subplots(nrows=3,ncols=2)
```

```python
# now lets do one example with our plot: 

# I want to use two axes one for all devs, the other one for Python devs and Java devs
fig1, (ax1,ax2) = plt.subplots(nrows=2, ncols=1)

# ax1 will be for all devs and ax2 will be for Python devs and Java devs

ax1.plot(age, all_devs, label='All devs salary')

ax2.plot(age, py,color='red', label='Python dev salary')

ax2.plot(age, jv, linestyle='--', label='JavaScript Salary')

ax1.set_title('Median Salary for all Devs')

ax1.set_ylabel('Salary')
ax1.legend()

# fig2 


ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary')
ax2.legend()


plt.tight_layout()
plt.show()

# we just ploted 2 axis on the same figure 
```

```python
# we can aslo use two different figures... We just need to add a figure at the beginning 
```

```python
# lets create 2 figures instead of two plots on one figure, like the example above.

# no need to add nrows and ncols since we're just plotting one axis on each figure

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()



ax1.plot(age, all_devs, label='All devs salary')

ax2.plot(age, py,color='red', label='Python dev salary')

ax2.plot(age, jv, linestyle='--', label='JavaScript Salary')

ax1.set_title('Median Salary for all Devs')

ax1.set_ylabel('Salary')
ax1.legend()

# fig2 


ax2.set_xlabel('Ages')
ax2.set_ylabel('Salary')
ax2.legend()


plt.tight_layout()
plt.show()

# this prints out two different figures. 
```

```python
# saving figures

# once you're done working on plot you can save them:

fig1.savefig('fig1.png')

fig2.savefig('fig2.png')


# you can automate this process so it sends you an email of the plot after saving it as a .png, that way you dont have to jump into the code to look at the figures

# another idea (Working with Live Data) : 
#  Write a python script that pulls data from an API
#  Save that data as a CSV file 
#  write another script that reads the data and analyses it and plots the results on different figures
#  Have the script save the figures as .png and the save them in a folder or send them to your email.
# this whole process can run continuously without user input.
```
