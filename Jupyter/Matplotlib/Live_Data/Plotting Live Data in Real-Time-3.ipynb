```python
%matplotlib notebook # helps with plotting live data in jupyter
import random
from itertools import count
from matplotlib import pyplot as plt 
import pandas as pd
from matplotlib.animation import FuncAnimation
```

```python
# style of my plots:

plt.style.use('ggplot')
```

```python


x_vals=[]

y_vals=[]

index= count()

# create a function for the animation:
def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    plt.cla() #clears out the axis after each plot
    plt.plot(x_vals, y_vals)
    

an= FuncAnimation(plt.gcf(),animate,interval=1000)  
#1st parameter: get current figure(plt.plot), 
#2: function we're running for our animation
#3: the interval for how often we want to run this function (2000= 2secs)



plt.tight_layout()
plt.show()
```

```python

# create a function for the animation:
def animate(i):
    
    data= pd.read_csv('live_data.csv') #This is live data from a python script that  continuously add rows
    
    x= data['x_value']

    y1= data['total_1']

    y2= data['total_2']
    
    plt.cla() #clears out the axis after each plot
    
    plt.plot(x, y1,label='y1')
    plt.plot(x, y2,label='y2')
    
    plt.legend(loc='upper right')
    
    

an= FuncAnimation(plt.gcf(),animate,interval=1000)  
#1st parameter: get current figure(plt.plot), 
#2: function we're running for our animation
#3: the interval for how often we want to run this function (2000= 2secs)



plt.tight_layout()
plt.show()
```
