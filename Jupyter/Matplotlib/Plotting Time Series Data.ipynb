```python
import pandas as pd
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
dates = [
    datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
]
```

```python
plt.style.use('bmh')
```

```python
y = [0, 1, 3, 4, 6, 5, 7]
```

```python
plt.plot_date(dates,y,linestyle='solid')

plt.show()
```

```python
plt.plot_date(dates,y,linestyle='solid')

#dates are not visible, an auto format would help:
#(gcf==get current figure)

plt.gcf().autofmt_xdate()

plt.show()
```

```python


plt.plot_date(dates,y,linestyle='solid')


plt.gcf().autofmt_xdate()

#changed the format of date from yyyy-mm-dd to 
date_format = mpl_dates.DateFormatter('%b, %d %Y')

#gca == get current axis

plt.gca().xaxis.set_major_formatter(date_format)
plt.show()

```

```python
data=pd.read_csv('Data/bitcoin.csv')

# importing my data but I need to make sure I import Dates in date format not strings:

data['Date'] = pd.to_datetime(data['Date'])

data.sort_values('Date',inplace=True)
```

```python
data # some random Bitcoin prices
```

```python
#defining my variables 
price_close = data['Close']

price_date = data['Date']
```

```python
plt.title('Bitcoin Prices')

plt.xlabel('Date')

plt.ylabel('Closing price')

plt.plot_date(price_date, price_close, linestyle='solid')

plt.gcf().autofmt_xdate()

plt.show()
```
